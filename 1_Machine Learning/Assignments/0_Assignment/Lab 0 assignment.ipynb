# Hands-on exercise

In this simple example you are required to perform a simple linear regression with scipy. Find all the information on the function in the documentation: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html

```python
import matplotlib.pyplot as plt
import csv
from scipy import stats
import numpy as np
import sklearn as sl
from sklearn import linear_model
```

```python
# Load the provided .csv file with the used car data

filename = "data/km_year_power_price.csv"
lines = csv.reader(open(filename, newline=''), delimiter=',')
next(lines)
# place your loading code here
km_list = []
year_list = []
power_list = []
price_list = []

for line in lines:
    km_list.append(int(line[0]))
    year_list.append(float(line[1]))
    power_list.append(float(line[2]))
    price_list.append(float(line[3]))
    
km_list = np.array(km_list)
year_list = np.array(year_list)
power_list = np.array(power_list)
price_list = np.array(price_list)
```

Use a linear regression to estimate the car prices from the year, kilometers or engine power. 
You can make a simple 1D regression from each one of the parameters independently 




```python
# linear regression with linregress (estimate price from year)
# your code.....
slope_a, intercept_a, r_value_a, p_value_a, std_err_a = stats.linregress(year_list,price_list)
x = np.linspace(min(year_list),max(year_list))
plt.show()
plt.plot(year_list,price_list,'o')
plt.plot(x,intercept_a+slope_a*x)
```

```python
# (Optional) linear regression with linear_model.LinearRegression() (estimate price from year)
# your code.....
reg = linear_model.LinearRegression().fit(year_list.reshape(-1,1),price_list)
x = np.linspace(min(year_list),max(year_list))
plt.show()
plt.plot(year_list,price_list,'o')
plt.plot(x,reg.intercept_+reg.coef_*x)
```

```python
# linear regression with least squares (estimate price from year)
# You should get exactly the same solution of linregress !
# your code.....
x = np.array([[1,i] for i in year_list])
b = np.dot(np.matrix.transpose(x),price_list)
A = np.dot(np.matrix.transpose(x),x)
w = np.dot(np.linalg.inv(A),b)
x = np.linspace(min(year_list),max(year_list))
plt.show()
plt.plot(year_list,price_list,'o')
plt.plot(x,w[0]+w[1]*x)
```

```python
# Plot the data and the lines representing the output of the linregress and least square algorithms
# your code.....
x = np.linspace(min(year_list),max(year_list))
plt.show()
plt.plot(year_list,price_list,'o')
plt.plot(x,w[0]+w[1]*x) #least square alg
plt.plot(x,intercept_a+slope_a*x) #linregress
plt.plot(x,reg.intercept_+reg.coef_*x) #linear_model.LinearRegression()
```

```python
# linear regression with linregress (estimate price from power)
# your code.....
slope_b, intercept_b, r_value_b, p_value_b, std_err_b = stats.linregress(power_list,price_list)
x = np.linspace(min(power_list),max(power_list))
plt.show()
plt.plot(power_list,price_list,'o')
plt.plot(x,intercept_b+slope_b*x)
```

```python
# linear regression with linregress (estimate price from km)
# your code.....
slope_c, intercept_c, r_value_c, p_value_c, std_err_c = stats.linregress(km_list,price_list)
x = np.linspace(min(km_list),max(km_list))
plt.show()
plt.plot(km_list,price_list,'o')
plt.plot(x,intercept_c+slope_c*x)
```

```python
# Have a look at the correlation coefficients to see which of the 3 features works better
# your code......
print(r_value_a,r_value_b,r_value_c)
print('It works better with power!')
```
