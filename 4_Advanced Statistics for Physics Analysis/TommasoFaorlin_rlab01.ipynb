## Exercise 1 - vectors and data frames

```r
#vectors for lake's name and parameters
Loch <- c('Loch Ness', 'Loch Lomond', 'Loch Morar', 'Loch Tay',
          'Loch Awe', 'Loch Maree', 'Loch Ericht', 'Locht Lochy',
          'Loch Rannoch', 'Loch Shiel', 'Loch Katrine', 'Loch Arkaig',
          'Loch Shin' )
volume <- c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35)
area <- c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5)
lenght <- c(39, 36, 18.8, 23, 41, 20, 23, 16, 15.7, 28, 12.9, 19.3, 27.8)
max_depth <- c(230, 190, 310, 150, 94, 114, 156, 162, 134, 128, 151, 109, 49)
mean_depth <- c(132, 37, 87, 60.6, 32, 38, 57.6, 70, 51, 40, 43.4, 46.5, 15.5)
```

```r
#dataframe from vectors
scottish.lakes <- data.frame(row.names = Loch, volume, area, lenght, max_depth, mean_depth)
scottish.lakes
```

```r
#highest and lowest volume
min_vol = paste("Min volume:", min(scottish.lakes['volume']), "km^3 -", Loch[which.min(scottish.lakes$volume)])
max_vol = paste("Max volume:", max(scottish.lakes['volume']), "km^3 -", Loch[which.max(scottish.lakes$volume)])
#highest and lowest area
min_are = paste("Min area:", min(scottish.lakes['area']), "km^2 -", Loch[which.min(scottish.lakes$area)])
max_are = paste("Max area:", max(scottish.lakes['area']), "km^2 -", Loch[which.max(scottish.lakes$area)])
#prints
print(min_vol)
print(max_vol)
print(min_are)
print(max_are)
```

```r
#ordering data frame with respect to area and determine two largest areas

#lakes.sorted = scottish.lakes[order(-area),] or, better,
scottish.lakes.sorted = scottish.lakes[order(area, decreasing=TRUE),]

#a dataframe element is a list so I access it with [[i]]
max_are1 = paste("Min area:", scottish.lakes.sorted[['area']][1], "km^2 -", row.names(scottish.lakes.sorted)[1])
max_are2 = paste("Max area:", scottish.lakes.sorted[['area']][2], "km^2 -", row.names(scottish.lakes.sorted)[2])
#prints
print(max_are1)
print(max_are2)
```

```r
#area occupied by lakes
lakes_covering = paste("The total area occupied by lakes is:", sum(scottish.lakes$area), "km^2")
print(lakes_covering)
```

```r
coal <- tibble(read_csv('coal-production-by-country.csv'))
head(coal)
```

```r
string1 = paste("There are", length(unique(coal$Entity)), "different countries")
print(string1)
```

```r
barplot(table(coal$Entity), col='green3')
```

It is kinda hard to read this plot. Furthermore, inside the raw dataset some artifacts can be found (for example there is a voice 'World')

```r
#selecting only rows of data after 1970
#I could have also used filter()
coal_g1970 <- coal[coal['Year']>1970,]

#I define my own 'not in'
'%!in%' <- function(x,y)!('%in%'(x,y))

excluded = c('Africa', 'EU-28','CIS','World','Asia and Oceania', 'Asia Pacific', 'OECD', 'North America', 'Eurasia', 'Europe')
coal_g1970_clean <- coal_g1970[coal_g1970$Entity%!in%excluded,]

int_prod <- aggregate(coal_g1970_clean$`Coal production (TWh)`, list(coal_g1970_clean$Entity), sum)
int_prod <- int_prod[order(int_prod$x, decreasing=TRUE),]
int_prod_top5 <- int_prod[1:5,]
int_prod_top5
```

```r
coal_top5 <- subset(coal_g1970_clean, coal_g1970_clean$Entity %in% int_prod_top5$Group.1)
```

```r
ggplot(coal_top5, aes(Year,`Coal production (TWh)`, group=factor(Entity), color=factor(Entity))) + 
    geom_line(size=2) +
    labs(x="Year", y="Coal Production (TWh)", color="'Big 5' countries")+
    ggtitle("Top 5 countries in coal production")
```

```r
ggplot(coal_g1970 %>% subset(Entity == "World") %>% mutate(CumSum = cumsum(`Coal production (TWh)`))) + 
aes (x = Year, y = CumSum, colour = Entity) +
geom_line(size =2) +
labs(x='Year', y='Coal Production (TWh)', color = 'Legend')+
ggtitle('World cumulative coal production')
```

## Exercise 4 - Covid19 Vaccine Data

### Italy

```r
vaccines <- tibble(read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations-by-manufacturer.csv'))
vaccines.ita = filter(vaccines, location=='Italy' )

plt1 <- ggplot(vaccines.ita, aes(date,total_vaccinations, group=factor(vaccine), color=factor(vaccine)))  +
        geom_line(size=1) +
        ggtitle("Vaccines per day for each company in Italy") +
        labs(x="Date", y="Total vaccinations", color="Companies") +
        geom_point(shape=2)

vaccines.ita.sum <- aggregate(vaccines.ita$total_vaccinations, by=list(date=vaccines.ita$date), sum)
vaccines.ita.sum['cases'] <- c(NA,diff(vaccines.ita.sum[[2]]))

plt2 <- ggplot(vaccines.ita.sum, aes(date, cases)) +
        geom_line(size=1) +
        ggtitle("Vaccine shots per day in Italy") +
        labs(x="Date", y="Total vaccinations") +
        geom_point(shape=2)

plt1
plt2
```
